```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
X = np.linspace(0, 10, 20).reshape(-1, 1)
y = 2 + 1.5 * X - 0.3 * X**2 + np.random.randn(20, 1) * 2   # رابطه غیرخطی با نویز
```

```python
plt.scatter(X, y, color="black") 
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)
```

```python
plt.scatter(X, y, color="black", label="real data")       # داده‌های واقعی
plt.plot(X, y_pred_lin, color="blue", label="Linear") # مدل خطی
plt.show()
```

```python
# --- مدل غیرخطی (Polynomial Regression) ---
poly = PolynomialFeatures(degree=3)   # اضافه کردن x^2
X_poly = poly.fit_transform(X)

poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_pred_poly = poly_reg.predict(X_poly)
```

```python
plt.scatter(X, y, color="black", label="real data")       # داده‌های واقعی
plt.plot(X, y_pred_poly, color="blue", label=" non Linear") # مدل  غیر خطی
plt.show()
```
