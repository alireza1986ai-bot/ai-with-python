```python
# داده‌های مربوط به دانش‌آموزان
data = {
    'Name': ['Ali', 'Sara', 'Reza', 'Mina', 'Hassan','nima','kamran'],
    'Age': [17, 16, 18, 17, 16, 17,19],
    'Math_Score': [85, 90, 78, 92, 88,96,72],
    'City': ['Tehran', 'Shiraz', 'Isfahan', 'Tabriz', 'Mashhad','Tehran','Tabriz']
}

# ایجاد DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
# نمایش اطلاعات کلی دیتافرم
print(df.info())
```

```python
# نمایش آماری از داده‌ها
print(df.describe())
```
