```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# داده‌های نمونه (ساعت مطالعه و نمره)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)  # ویژگی‌ها
y = np.array([2, 4, 5, 4, 5, 7])  # برچسب‌ها
```

```python

plt.scatter(X, y, color="blue")
plt.xlabel("time")
plt.ylabel("Score")
plt.title("Data")
plt.show()
```

```python
model = LinearRegression()

# آموزش مدل
model.fit(X, y)

```

```python
# نمایش ضرایب
print("ضریب (وزن):", model.coef_)
print("عرض از مبدأ:", model.intercept_) 
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
y 
```

```python
plt.scatter(X, y, color="blue", label="real  data")
plt.plot(X, y_pred, color="red", label="LinearRegression")
plt.xlabel("time")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MSE:", mse)
print("R^2:", r2)
```
