```python

# هزینه‌های تبلیغات و تعداد فروش‌ها
advertising_costs = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60]  # هزینه‌های تبلیغات به هزار دلار
online_sales = [100, 120, 150, 180, 210, 240, 270, 300, 330, 360]  # تعداد فروش‌های آنلاین

# رسم نمودار پراکندگی
plt.scatter(advertising_costs, online_sales, color='green', marker='o')

# تنظیمات نمودار
plt.xlabel("Advertising Costs (in thousand dollars)")
plt.ylabel("Online Sales (units)")
plt.title("Scatter Plot of Advertising Costs vs. Online Sales")

# نمایش نمودار
plt.show()
```

```python
## import matplotlib.pyplot as plt

gdp = [2000, 2100, 2200, 2300, 2400, 2500, 2600, 2700, 2800, 2900]  # تولید ناخالص داخلی به میلیارد دلار
unemployment_rate = [5.0, 4.8, 4.7, 4.5, 4.3, 4.2, 4.0, 3.8, 3.7, 3.5]  # نرخ بیکاری به درصد

# اندازه و رنگ نقاط
sizes = [50, 70, 90, 110, 130, 150, 170, 190, 210, 230]  # اندازه نقاط (مثلاً حجم بازار)
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'black', 'pink']  # رنگ نقاط (مثلاً تغییرات فصلی)

# رسم نمودار پراکندگی با تنظیمات سفارشی
plt.scatter(gdp, unemployment_rate, s=sizes, c=colors, marker='o', alpha=0.9)

# تنظیمات نمودار
plt.xlabel("GDP (in billion dollars)")
plt.ylabel("Unemployment Rate (percent)")
plt.title("Scatter Plot of GDP vs. Unemployment Rate")

# نمایش نمودار
plt.show()
```
