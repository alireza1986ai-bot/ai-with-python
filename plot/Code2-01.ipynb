```python
##### # داده‌های مالی: درآمد ماهانه یک شرکت (در هزار دلار)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']
revenue = [120, 135, 145, 150, 160, 170, 180]

# داده‌های اجتماعی: تعداد فالوورها (در هزار نفر)
followers = [50, 55, 60, 62, 68, 70, 75]

# رسم نمودار درآمد
plt.plot(months, revenue, marker='*', linestyle='-', color='blue', label='Revenue', linewidth=.5)

# رسم نمودار تعداد فالوورها
plt.plot(months, followers, marker='o', linestyle='--', color='red', label='Followers', linewidth=2)

# تنظیمات محور x و y
plt.xlabel("Months")
plt.ylabel("Value")

# عنوان نمودار
plt.title("Monthly Revenue and Social Media Followers")

# افزودن راهنما (legend) برای شناسایی خطوط
plt.legend()

# نمایش نمودار
plt.show()
```

```python
import numpy as np
x=np.linspace(0,7,200)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,7,50)
y=np.cos(x)
plt.plot(x,y)
plt.xlim(-3,9)
plt.ylim(-2,3)
plt.show()
```

```python
x=np.linspace(0,7,50)
y=np.cos(x)
plt.plot(x,y)
plt.axis([-3,9,-2,3])
plt.show()
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
y=np.sin(x)
plt.plot(x,y,color='blue')
plt.xticks(np.linspace(-6,6,9))
plt.yticks([-1,-0.5, 0,0.5,1])
plt.show()
```
