```python
import matplotlib.pyplot as plt
```

```python


# داده‌ها
companies = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Others']
market_share = [31, 28, 14, 10, 17]

# رسم نمودار دایره‌ای
plt.pie(market_share, labels=companies, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple', 'orange'])

# تنظیمات نمودار
plt.title("Smartphone Market Share (2020)")

# نمایش نمودار
plt.show()
```

```python
import matplotlib.pyplot as plt

# داده‌ها
categories = ['R&D', 'Marketing', 'Operations', 'Sales', 'Others']
budget = [4, 25, 20, 10, 5]

explode = (0, 0, .1, 0, 0)  # اولین بخش کمی بیرون‌تر نمایش داده می‌شود

# رسم نمودار دایره‌ای سه‌بعدی
plt.pie(budget, labels=categories, autopct='%2.1f%%',  explode=explode, colors=['cyan', 'magenta', 'yellow', 'blue', 'grey'], shadow=True)

# تنظیمات نمودار
plt.title("3D Effect Pie Chart")

# نمایش نمودار
plt.show()
```
