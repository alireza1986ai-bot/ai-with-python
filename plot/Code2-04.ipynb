```python
import matplotlib.pyplot as plt
```

```python


# داده‌ها
countries = ['USA', 'China', 'Japan', 'Germany', 'India']
gdp = [15035, 14343, 5082, 3846, 2875]  # تولید ناخالص داخلی به میلیارد دلار

# رسم نمودار میله‌ای
plt.bar(countries, gdp, color='green')

# تنظیمات نمودار
plt.xlabel("Countries")
plt.ylabel("GDP in Billions USD")
plt.title("GDP of Different Countries (2020)")

# نمایش نمودار
plt.show()
```

```python
import matplotlib.pyplot as plt

# داده‌های اقتصادی
sectors = ['Technology', 'Healthcare', 'Finance', 'Energy', 'Consumer Goods']
revenue = [1200, 850, 950, 700, 600]  # درآمد سالانه به میلیون دلار
colors = ['blue', 'green', 'orange', 'red', 'purple']  # رنگ‌های مختلف برای هر بخش

# رسم نمودار میله‌ای افقی
plt.barh(sectors, revenue, color=colors)

# تنظیمات نمودار
plt.xlabel("Revenue (in million dollars)")
plt.ylabel("Sectors")
plt.title("Annual Revenue by Industry Sector")

# نمایش نمودار
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# داده‌های اجتماعی
categories = ['Photos', 'Videos', 'Articles', 'Stories', 'Comments']
posts_network1 = [150, 200, 100, 80, 120]  # تعداد پست‌ها در شبکه اجتماعی 1
posts_network2 = [120, 180, 110, 90, 130]  # تعداد پست‌ها در شبکه اجتماعی 2

# تنظیمات برای گروه‌بندی میله‌ها
bar_width = 0.35
index = np.arange(len(categories))

# رسم نمودار میله‌ای گروه‌بندی شده
plt.bar(index, posts_network1, bar_width, color='blue', label='Network 1')
plt.bar(index + bar_width, posts_network2, bar_width, color='red', label='Network 2')

# تنظیمات نمودار
plt.xlabel("Post Categories")
plt.ylabel("Number of Posts")
plt.title("Number of Posts by Category in Two Social Networks")
plt.xticks(index + bar_width / 2, categories)
plt.legend()

# نمایش نمودار
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# داده‌های اجتماعی
categories = ['Photos', 'Videos', 'Articles', 'Stories', 'Comments']
posts_network1 = [150, 200, 100, 80, 120]  # تعداد پست‌ها در شبکه اجتماعی 1
posts_network2 = [120, 180, 110, 90, 130]  # تعداد پست‌ها در شبکه اجتماعی 2

# تنظیمات برای گروه‌بندی میله‌ها
bar_width = 0.35
index = np.arange(len(categories))

# رسم نمودار میله‌ای گروه‌بندی شده
plt.bar(categories, posts_network1, bar_width, color='blue', label='Network 1')
plt.bar(categories, posts_network2, bar_width, color='red', label='Network 2')

# تنظیمات نمودار
plt.xlabel("Post Categories")
plt.ylabel("Number of Posts")
plt.title("Number of Posts by Category in Two Social Networks")
plt.xticks(index + bar_width / 2, categories)
plt.legend()

# نمایش نمودار
plt.show()
```
