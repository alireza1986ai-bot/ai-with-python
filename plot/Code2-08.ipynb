```python

# ایجاد داده‌های اجتماعی: تعداد لایک‌های پست‌ها
np.random.seed(0)
likes = np.random.normal(loc=150, scale=30, size=100)  # تعداد لایک‌ها به طور تصادفی

# رسم نمودار توزیع
sns.histplot(likes, color='green') 

# تنظیمات نمودار
plt.xlabel('Number of Likes')
plt.ylabel('Frequency')
plt.title('Distribution Plot of Likes on Social Media Posts')

# نمایش نمودار
plt.show()
```

```python
# ایجاد داده‌های دستی برای گروه‌های مختلف
data = pd.DataFrame({
    'Category': ['Group A'] * 10 + ['Group B'] * 10 + ['Group C'] * 10,
    'Value': [10, 12, 11, 14, 15, 13, 12, 11, 15, 14,  # Group A
              20, 22, 21, 19, 25, 24, 23, 21, 22, 23,  # Group B
              30, 32, 31, 29, 34, 33, 32, 30, 31, 29]   # Group C
})


print(data)
```

```python
# رسم نمودار جعبه‌ای
sns.boxplot(x='Category', y='Value', data=data, palette='Set2')

# تنظیمات نمودار
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Box Plot of Values by Category')

# نمایش نمودار
plt.show()
```

```python

# داده‌های اقتصادی
countries = ['USA', 'China', 'Germany', 'India', 'Brazil']
years = [2020, 2021, 2022, 2023, 2024]
gdp_data = {
    'USA': [21100, 22300, 23400, 24500, 25600],
    'China': [14100, 15400, 16700, 18000, 19300],
    'Germany': [3800, 4000, 4200, 4400, 4600],
    'India': [2700, 2900, 3100, 3300, 3500],
    'Brazil': [2100, 2200, 2300, 2400, 477500]
}

# ایجاد DataFrame
df = pd.DataFrame(gdp_data, index=years).T

print(df)
```

```python
# رسم نمودار حرارتی
plt.figure(figsize=(12, 6))
sns.heatmap(df, annot=False, cmap='YlGnBu', fmt='d', linewidths=0.5)

# تنظیمات نمودار
plt.xlabel('Year')
plt.ylabel('Country')
plt.title('GDP of Different Countries Over Years')

# نمایش نمودار
plt.show()


```
