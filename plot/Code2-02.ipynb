```python
import matplotlib.pyplot as plt
```

```python
# انتخاب سبک
plt.style.use('ggplot')
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul']
revenue = [120, 135, 145, 150, 160, 170, 180]
followers = [50, 55, 60, 62, 68, 70, 75]
plt.plot(months, revenue, marker='*', linestyle='-', color='blue', label='Revenue', linewidth=1)
plt.plot(months, followers, marker='o', linestyle='--', color='red', label='Followers', linewidth=2)
plt.xlabel("Months")
plt.ylabel("Value")
plt.title("Monthly Revenue and Social Media Followers")
plt.legend()
plt.show()
```
